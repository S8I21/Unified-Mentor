```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/SHIVANSH/Downloads/data.csv')
```

# PERCENTAGE OF PEOPLE LEFT

```python
# Calculate the value counts of Attrition (1 for left, 0 for stayed)
attrition_counts = df['Attrition'].value_counts()

# Rename the index for clarity
attrition_counts.index = ['Stayed', 'Left']
```

```python


# Plot the pie chart
attrition_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])

plt.title('Attrition Distribution - Stayed vs. Left')
plt.ylabel('')  # Hide the y-label
plt.show()
```

# MARTIAL STATUS VS ATTRTION

```python

# Filter the data for those who have left
left_df = df[df['Attrition'] == 1]

# Filter the data for those who have not left
not_left_df = df[df['Attrition'] == 0]


# Pie chart for those who have left
left_marital_status_counts = left_df['MaritalStatus'].value_counts()
left_marital_status_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('Attrition according to Martial Status - Left')
plt.ylabel('')  # Hide the y-label
plt.show()

# Pie chart for those who have not left
not_left_marital_status_counts = not_left_df['MaritalStatus'].value_counts()
not_left_marital_status_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('Attrition by Marital Status - Not Left')
plt.ylabel('')  # Hide the y-label
plt.show()
```

# GENDER VS ATTRITION


```python
# Filter data for males
male_df = df[df['Gender'] == 'Male']

# Filter data for females
female_df = df[df['Gender'] == 'Female']
```

```python
# Pie chart for male attrition
male_attrition_counts = male_df['Attrition'].value_counts()
male_attrition_counts.index = ['Not Left', 'Left']  # Rename the index for clarity
male_attrition_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])

plt.title('Attrition by Gender - Male')
plt.ylabel('')  # Hide the y-label
plt.show()

# Pie chart for female attrition
female_attrition_counts = female_df['Attrition'].value_counts()
female_attrition_counts.index = ['Not Left', 'Left']  # Rename the index for clarity
female_attrition_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])

plt.title('Attrition by Gender - Female')
plt.ylabel('')  # Hide the y-label
plt.show()
```

# BUSINESS TRAVEL VS ATTRITION

```python
# Create a crosstab of BusinessTravel vs Attrition
business_travel_attrition = pd.crosstab(df['BusinessTravel'], df['Attrition'])

# Plot the stacked bar chart
business_travel_attrition.plot(kind='bar', stacked=True, color=['#66b3ff','#ff9999'])

plt.title('Attrition by Business Travel')
plt.xlabel('Business Travel')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)  # Rotate x labels if needed
plt.legend(['Stayed', 'Left'], loc='upper right')
plt.show()
```

# IMPACT OF DISTANCE

```python


# Use 'hue' to assign colors to the Attrition categories
sns.boxplot(x='Attrition', y='DistanceFromHome', hue='Attrition', data=df, palette=['#66b3ff','#ff9999'], dodge=False)

plt.title('Impact of Travel Distance on Attrition')
plt.xlabel('Attrition')
plt.ylabel('Distance from Home')
plt.xticks([0, 1], ['Stayed', 'Left'])
plt.legend([],[], frameon=False)  # Remove the redundant legend
plt.show()
```

```python
# Scatter plot of Age vs MonthlyIncome, colored by Attrition
sns.scatterplot(x='Age', y='MonthlyIncome', hue='Attrition', data=df, palette=['#66b3ff','#ff9999'])

plt.title('Age vs Monthly Income - Impact on Attrition')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.legend(['Stayed', 'Left'])
plt.show()
```

# Fair Income Distribution

```python
bins = [18, 20, 30, 40, 50, 60]
labels = ['18-20', '21-30', '31-40', '41-50', '51-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Set the style for the plots
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axs = plt.subplots(3, 1, figsize=(14, 18))

# Boxplot for Monthly Income based on Education Level
sns.boxplot(x='AgeGroup', y='MonthlyIncome', hue='Education', data=df, ax=axs[0])
axs[0].set_title('Monthly Income based on Education Level')
axs[0].set_xlabel('Age Group')
axs[0].set_ylabel('Monthly Income')

# Boxplot for Monthly Income based on Tenure
sns.boxplot(x='AgeGroup', y='MonthlyIncome', hue='Tenure', data=df, ax=axs[1])
axs[1].set_title('Monthly Income based on Tenure')
axs[1].set_xlabel('Age Group')
axs[1].set_ylabel('Monthly Income')

# Boxplot for Monthly Income based on Training Times Last Year
sns.boxplot(x='AgeGroup', y='MonthlyIncome', hue='TrainingTimesLastYear', data=df, ax=axs[2])
axs[2].set_title('Monthly Income based on Training Times Last Year')
axs[2].set_xlabel('Age Group')
axs[2].set_ylabel('Monthly Income')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
attrition_data = df[df['Attrition'] == 1]

# Count the number of attritions in each education field
attrition_counts = attrition_data['EducationField'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Attrition by Education Field')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
